**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

Chosen method: Two sample T test

```python
pokemons_dragons = df[df['Type 1']=='Dragon']['HP']
pokemons_grass = df[df['Type 1'] == 'Grass']['HP']
```

```python
#HO avg hp dragons <= avg hp grass
#H1 avg hp dragons > avg hp grass

#sl= 0.05

st.ttest_ind(pokemons_dragons, pokemons_grass, equal_var=False, alternative='greater')

#pvalue > sl meaning that we have enough evidences to do not reject H0
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


Chosen method: ANOVA

```python
#HO: legendary pokekons stats = non legendary pokemons stats
#H1: legendary pokemons stats != non legendary pokemons stats
#sl = 0.05

legendary_pokemons = df[df['Legendary']==True][['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
non_legendary_pokemons = df[df['Legendary']==False][['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

st.f_oneway(legendary_pokemons, non_legendary_pokemons)

#H0 rejected, stays H1, we have enough evidences to say that legendery pokemons != non legendary in terms of all the stats that we presented
```

In this challenge, we will be working with california-housing data. The data can be found here:

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
import math
```

```python
def euclidean_distance(lat1, lon1, lat2, lon2):
    dx = lon2 - lon1
    dy = lat2 - lat1
    return math.sqrt(dx**2 + dy**2)
```

```python
school = (-118, 34)
hospital = (-122, 37)
df['distance_from_school'] = df.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], school[0], school[1]), axis=1).round(2)
df['distance_from_hospital'] = df.apply(lambda row: euclidean_distance(row['longitude'], row['latitude'], hospital[0], hospital[1]), axis=1).round(2)
    
```

```python
df.head()
```

```python
houses_close_to_hospital_school = df[(df['distance_from_hospital']<=0.5) | (df['distance_from_school']<=0.5)]['median_house_value']
houses_far_from_hospital_school = df[(df['distance_from_hospital']>0.5) & (df['distance_from_school']>0.5)]['median_house_value']
```

```python
#HO: houses_close_to_hospital_school avg price <= houses_far_from_hospital_school avg price
#H1: houses_close_to_hospital_school avg price > houses_far_from_hospital_school avg price

#sl = 0.05

st.ttest_ind(houses_close_to_hospital_school, houses_far_from_hospital_school, equal_var=False, alternative='greater')
#we reject H0, we have collected enough evidences to keep the H1 and reject H0 as the p_value is lower than sl
```
